## **Stage 3 – EDA and Visualizations**





The chart shows the relationship between a game’s average rating and its recent growth on the Google Play Store.
Each point represents a game, with the average rating on one axis and 30-day growth on the other.

Overall, the chart indicates that games with higher average ratings tend to have higher growth, while lower-rated games generally show slower growth. This suggests that user satisfaction (ratings) plays an important role in driving a game’s popularity and growth.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

**Chart 1: Distribution of Average Ratings (Histogram)**

**Explanation:**
This chart shows how average ratings are distributed across all games.
Most games are clustered between 4 and 5 stars, indicating that the majority of games in the dataset are highly rated by users. Very few games fall into lower rating ranges.

**Insight:**
High user satisfaction is common among top-rated games.

```python
plt.figure()
plt.hist(df['average rating'], bins=10, color='teal', edgecolor='black')
plt.xlabel("Average Rating")
plt.ylabel("Number of Games")
plt.title("Distribution of Average Ratings")
plt.show()
```

**Chart 2: Top 10 Games by Total Ratings (Horizontal Bar Chart)**

**Explanation:**
This chart highlights the top 10 most-rated games on the Play Store.
A small number of games receive a very high number of ratings compared to others.

**Insight:**
The market is highly competitive, with a few games dominating user attention.

```python
top10 = df.sort_values("total ratings", ascending=False).head(10)
plt.figure()
plt.barh(top10['title'], top10['total ratings'])
plt.xlabel("Total Ratings")
plt.ylabel("Game Title")
plt.title("Top 10 Games by Total Ratings")
plt.show()
```

**Chart 3: Star Rating Distribution (Bar Chart)**

**Explanation:**
This chart shows the average number of 1-star to 5-star ratings.
The dominance of 4-star and 5-star ratings indicates positive user feedback.

**Insight:**
Overall user sentiment toward these games is highly positive.

```python
stars = df[['5 star ratings','4 star ratings','3 star ratings',
            '2 star ratings','1 star ratings']].mean()

plt.figure()
stars.plot(kind='bar')
plt.xlabel("Star Rating Type")
plt.ylabel("Average Count")
plt.title("Average Star Rating Distribution")
plt.show()
```

**Chart 4: Average Rating by Category Count Plot – Game Categories**

**Explanation:**
This chart displays the average rating for each game category.
Some categories consistently achieve higher ratings, while others perform comparatively lower.

**Insight:**
Category selection plays a major role in determining game success.

```python
plt.figure(figsize=(10,5))
sns.countplot(x="category", data=df, palette="Set2")
plt.xticks(rotation=90)
plt.title("Count of Games by Category")
plt.show()
```

**Chart 5: Paid vs Free Games (Box Chart)**

**Explanation:**
This chart compares the number or average ratings of free and paid games.
Free games clearly dominate the dataset, showing that most highly rated games follow a free-to-play model.

**Insight:**
Free games attract more users due to a lower entry barrier.

```python
#Box Plot – Total Ratings (Free vs Paid)

plt.figure()
sns.boxplot(x="paid", y="total ratings", data=df, palette="pastel")
plt.yscale("log")
plt.title("Total Ratings by Paid/Free")
plt.show()
```

**Chart 6: 30-Day Growth vs 60-Day Growth (Scatter Plot)**

**Explanation:**
This chart compares short-term growth with medium-term growth.
Games with high 30-day growth often show high 60-day growth as well.

**Insight:**
Early growth is a strong indicator of future performance.

```python
plt.figure()
plt.plot(df['rank'], df['growth (30 days)'], label="30 Days")
plt.plot(df['rank'], df['growth (60 days)'], label="60 Days")
plt.xlabel("Rank")
plt.ylabel("Growth")
plt.title("30 vs 60 Days Growth")
plt.legend()
plt.show()
```

**Chart 7: Heatmap – Pivot Table (Category vs Paid)**

**Insights**:The heatmap shows that across all game categories, free games are significantly more than paid games.

 Very few paid games appear in any category, indicating that the free-to-play model dominates across all genres.

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

**Chart 8: Total Rating VS Average Rating**

---



**Explanation:**
This chart tell comparison of totalrating vs average rating.

**Insight:**
rating is a strong indicator of future performance.

```python
plt.figure()
plt.scatter(df['total ratings'], df['average rating'])
plt.xlabel("Total Ratings")
plt.ylabel("Average Rating")
plt.title("Total Ratings vs Average Rating")
plt.show()
```

**Chart9: Violin Plot: Average Rating by Paid/Free**

**Insight**: The violin plot compares the distribution of average ratings between free and paid games.

```python
# 9. Violin Plot: Average Rating by Paid/Free
plt.figure()
sns.violinplot(x="paid", y="average rating", data=df, palette="Set2")
plt.title("Average Rating by Paid vs Free")
plt.xlabel("Paid")
plt.ylabel("Average Rating")
plt.show()
```

**Chart 10: Distribution of Average Ratings**

**Insight**
The distribution of average ratings shows that most games have ratings.

```python
plt.figure()
sns.histplot(df['average rating'], bins=10, kde=True, color="green")
plt.title("Distribution of Average Ratings")
plt.xlabel("Average Rating")
plt.ylabel("Count")
plt.show()
```
